```python
import pandas as pd
```

```python
for column in df_stroke.columns:
    if column not in ["id", "age", "avg_glucose_level", "bmi"]:
        print(df_stroke[column].value_counts())
        print((df_stroke[column].value_counts() / len(df_stroke))*100)
```

```python
print(sum(pd.isna(df_stroke["smoking_status"])))
print((sum(pd.isna(df_stroke["smoking_status"])) / len(df_stroke))*100)
```
